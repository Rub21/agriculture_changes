```python
def display_multiple_img(images, rows = 1, cols=1, size=20,cmap='gist_earth'):
    figure, ax = plt.subplots(nrows=rows,ncols=cols,figsize=(size, size) )
    for ind,title in enumerate(images):
        ax.ravel()[ind].imshow(images[title].variable.data[0],cmap = cmap)
        ax.ravel()[ind].set_title(title)
        ax.ravel()[ind].set_axis_off()
    plt.tight_layout()
    plt.show()
```

```python
from sklearn import cluster

cl = cluster.KMeans(n_clusters=4) # create an object of the classifier
param = cl.fit(img1d) # train it
img_cl = cl.labels_ # get the labels of the classes
img_cl = img_cl.reshape(imgxyb[:,:,0].shape) # reshape labels to a 3d array (one band only)
```

# clustering

```python
import rasterio as rio
from rasterio.plot import show
from sklearn import cluster
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Open the image 
cosecha_raster = rio.open(data_path + "/cosecha_values.tif")

# create an empty array with same dimension and data type
empty_imgxyb = np.empty((cosecha_raster.height, cosecha_raster.width, cosecha_raster.count), cosecha_raster.meta['dtype'])
# loop through the raster's bands to fill the empty array
for band in range(empty_imgxyb.shape[2]):
    empty_imgxyb[:,:,band] = cosecha_raster.read(band+1)


# convert to 1d array the empty image
img1d=empty_imgxyb[:,:,:3].reshape((empty_imgxyb.shape[0]*empty_imgxyb.shape[1],empty_imgxyb.shape[2]))
```

```python
empty_imgxyb.shape
```

```python
# convert to 1d array
img1d=empty_imgxyb[:,:,:3].reshape((empty_imgxyb.shape[0]*empty_imgxyb.shape[1],empty_imgxyb.shape[2]))
img1d.shape
```

```python
cl = cluster.KMeans(n_clusters=4) # create an object of the classifier
param = cl.fit(img1d) # train it
img_cl = cl.labels_ # get the labels of the classes
img_cl = img_cl.reshape(empty_imgxyb[:,:,0].shape) # reshape labels to a 3d array (one band only)
```

```python
import matplotlib
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ["black","red","green","yellow"])
# Show the resulting array and save it as jpg image
plt.figure(figsize=[20,20])
plt.imshow(img_cl, cmap=cmap)
plt.axis('off')
plt.savefig("elhas_clustered.jpg", bbox_inches='tight')
plt.show()
```
